```python
import gzip
import math
import numpy
import random
import sklearn
import string
from collections import defaultdict
from gensim.models import Word2Vec
from nltk.stem.porter import *
from sklearn import linear_model
from sklearn.manifold import TSNE
import dateutil
import warnings
warnings.filterwarnings("ignore")
def assertFloat(x):
    assert type(float(x)) == float

def assertFloatList(items, N):
    assert len(items) == N
    assert [type(float(x)) for x in items] == [float]*N
dataset = []

f = gzip.open("young_adult_20000.json.gz")
for l in f:
    d = eval(l)
    dataset.append(d)
    if len(dataset) >= 20000:
        break
        
f.close()
answers = {}
```

```python
df = defaultdict(int)
tfall = []
for d in dataset[:10000]:
  tf = defaultdict(int)
  r = ''.join([c for c in d['review_text'].lower() if not c in punctuation])
  for w in r.split():
      df[w] += 1
  for w in r.split():
      tf[w] += 1
  tfall.append(tf)
```

```python
import numpy as np
from gensim import matutils  

def similarity_cosine(vec1, vec2):
    cosine_similarity = np.dot(matutils.unitvec(vec1), matutils.unitvec(vec2))
    return cosine_similarity
```

```python
wordCount = defaultdict(int)
punctuation = set(string.punctuation)
for d in dataset[:10000]:
  r = ''.join([c for c in d['review_text'].lower() if not c in punctuation])
  for w in r.split():
    wordCount[w] += 1

counts = [(wordCount[w], w) for w in wordCount]
counts.sort()
counts.reverse()

print(counts)
```

```python
words = [x[1] for x in counts[:1000]]
```

```python
tfidf = dict(zip(words,[tfall[0][w] * math.log2(len(dataset) / df[w]) for w in words]))
tfidfQuery = [tfall[0][w] * math.log2(len(dataset) / df[w]) for w in words]





maxTf = [(tfall[0][w],w) for w in words]
maxTf.sort(reverse=True)
maxTfIdf = [(tfidf[w],w) for w in words]
maxTfIdf.sort(reverse=True)
```

```python
def Cosine(x1,x2):
    numer = 0
    norm1 = 0
    norm2 = 0
    for a1,a2 in zip(x1,x2):
        numer += a1*a2
        norm1 += a1**2
        norm2 += a2**2
    if norm1*norm2:
        return numer / math.sqrt(norm1*norm2)
    return 0
```

```python
similarities = []
for i in range(1,10000):
  tfidf2 = [tfall[i][w] * math.log2(len(dataset) / df[w]) for w in words]
  similarities.append((Cosine(tfidfQuery, tfidf2), dataset[i]['review_text']))
similarities.sort(reverse=True)
```

#Q4

```python
usersPerItem = defaultdict(set) 
itemsPerUser = defaultdict(set) 
reviewsPerUser = defaultdict(list)
reviewsPerItem = defaultdict(list)
ratingDict = {} 
```

```python
for d in dataset: 
  itemsPerUser[d['user_id']].add(d['book_id'])
  usersPerItem[d['book_id']].add(d['user_id'])
  reviewsPerItem[d['book_id']].append(d['rating'])
  reviewsPerUser[d['user_id']].append(d['rating'])
  ratingDict[(d['user_id'],d['book_id'])] = d['rating']
```

```python
model10 = Word2Vec(reviewLists,
                 min_count=1, # Words/items with fewer instances are discarded
                 size=20, # Model dimensionality
                 window=4, # Window size
                 sg=1) # Skip-gram model
```

```python
globalrating = sum(ratingDict.values())/len(ratingDict)

def avgratingitem(item):
  if len(reviewsPerItem[item]) != 0 :
    return sum(reviewsPerItem[item])/len(reviewsPerItem[item])
  else:
    return 0

def predictRating(user,item):
  if item not in usersPerItem or user not in itemsPerUser:  
    return globalrating  
  num = 0
  denum = 0 
  for elem in itemsPerUser[user]:
    if elem != item:
      try: 
        num += ((ratingDict[(user,elem)]-avgratingitem(elem)) * model10.wv.distance(elem,item))
        denum += model10.wv.distance(elem,item)
      except: 
        if avgratingitem(elem) == 0 :
          return globalrating
        else:
          return avgratingitem(elem)
  if denum != 0:
    return avgratingitem(item) + num/denum
  return avgratingitem(item)
  

mse = 0
n = 0 
for item in dataset[:1000]:
  mse += (item['rating']-predictRating(item['user_id'],item['book_id']))**2
  n += 1
mse /= n

print(mse)

```

```python
answers["Q4"] = mse
answers
```

```python
PreMse=0
dataTest = dataset[:1000]
for i in dataTest:
  user = i['user_id']
  item = i["book_id"]
  y = i["rating"]
  y_pred = predictRating(user,item)
  if y_pred<0:
    y_pred = 0
  elif y_pred>5:
    y_pred = 5
  PreMse += (y-y_pred)**2

MSE = PreMse/len(dataTest)
print(MSE)
```
